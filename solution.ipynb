```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mkt_data = pd.read_csv('cleaned_mkt_data.csv').drop(columns = ['prueba'])
```

```python
mkt_data.info()
```

```python
mkt_data.describe().T
```

```python
fig = plt.figure(figsize = (12,8))
plt.title('Response: yes or no')
sns.countplot(x = 'response', data = mkt_data)
plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
plt.title('Response: yes or no - by channel')
sns.countplot(x = 'sales channel', hue = 'response', data = mkt_data)
plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
plt.title('total claim amount - by response')
sns.kdeplot(x = 'total claim amount', hue = 'response', data = mkt_data)
plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
plt.title('income - by response')
sns.kdeplot(x = 'income', hue = 'response', data = mkt_data)
plt.show()
```
