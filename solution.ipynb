```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mkt_data = pd.read_csv('cleaned_mkt_data.csv').drop(columns = ['prueba'])
```

```python
mkt_data.info()
```

```python
mkt_data.describe().T
```

```python
fig = plt.figure(figsize = (12,8))
plt.title('Total of negative responses vs positive responses', fontsize = 15)
sns.countplot(x = 'response', data = mkt_data)
plt.show()
```

```python
response_rate = mkt_data['response'].value_counts() / len(mkt_data)
response_rate = [response_rate[0], response_rate[1]]

fig = plt.figure(figsize = (12,8))
plt.bar(x = ['No', 'Yes'], height = response_rate)
plt.title('Response rate in % - no vs yes', fontsize = 15)
plt.ylabel('Percentage')
plt.xlabel('Response')
plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
plt.title('Response proportion: yes or no - by channel')
sns.countplot(x = 'sales channel', hue = 'response', data = mkt_data)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (15, 8))
axes[0].set_title('density of total claim amount per response')
sns.kdeplot(ax = axes[0], x = 'total claim amount', hue = 'response', data = mkt_data)
axes[1].set_title('barplot of total claim amount per response')
sns.boxplot(ax = axes[1], x = 'response', y = 'total claim amount', data = mkt_data)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (15, 8))
axes[0].set_title('density of income per response')
sns.kdeplot(ax = axes[0], x = 'income', hue = 'response', data = mkt_data)
axes[1].set_title('barplot of income per response')
sns.boxplot(ax = axes[1], x = 'response', y = 'income', data = mkt_data)
plt.show()
```
